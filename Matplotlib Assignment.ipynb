```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv")
```

2.4.1 Create a pie chart presenting the male/female proportion

```python
plt.pie(df['sex'].value_counts(),labels=('Male','Female'),autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')
```

2.4.2 Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```python
import seaborn as sns
sns.FacetGrid(df, hue="sex", height=6).map(plt.scatter, "fare", "age").add_legend()
```
